Q1. 2022-01-01 ~ 2022-12-01 월초 인덱스를 생성<br>
Q2. Q1에서 생성한 인덱스를 가지고 데이터프레임 생성(값은 자율, 이때 칼럼이름은 value)<br>
Q3. value 칼럼에서 100을 더한 value_add_100 칼럼을 생성해주세요.<br>
Q4. value 칼럼과 value_add_100 칼럼을 평균한 mean 칼럼을 생성해주세요.<br>

```python
#Q1.
import pandas as pd

dates = pd.date_range('2022-01-01','2022-12-31', freq = "MS")
dates
```

```python
#Q2.
value = {
    '서준범':[175,176,177,178,179,180,181,182,183,184,185,186]
}

df1 = pd.DataFrame(value, index = dates)
df1
```

```python
#Q3.
df1['pluesedvalue'] = df1['서준범'] + 100

df1
```

```python
#Q4.
df1['mean'] = (df1['서준범']+df1['pluesedvalue'])/2
df1
```
