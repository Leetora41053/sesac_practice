```python
#ax.plot & ax.scatter

import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)

n_data = 100

#np.random.normal(mean / std / size)순서로 작성
x_data = np.random.normal(0, 1, (n_data))
y_data = np.random.normal(0, 1, (n_data))

fig, ax = plt.subplots(figsize=(7, 7))
ax.scatter(x_data, y_data)

#ax.plot을 통해 그래프를 그릴 수 있고, 'o'를 통해 모양 설정 가능


ax.plot(x_data, y_data, 'o')
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)

x_min, x_max = -5, 5
n_data = 100

#low / high / size 순서로 uniform함수 사용
x_data = np.random.uniform(x_min, x_max, n_data)
y_data = x_data + 0.5*np.random.normal(0,1, n_data)

#범위에 따라 균등한 숫자를 출력하는 함수 np.linspace
pred_x = np.linspace(x_min, x_max, 2)
pred_y = pred_x

fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(x_data, y_data)
ax.plot(pred_x, pred_x, color='r', linewidth=3)
```

```python
#Size Array & Color Array

import matplotlib.pyplot as plt
import numpy as np

n_data = 10

x_data = np.linspace(0, 10, n_data)
y_data = np.linspace(0, 10, n_data)

s_arr = np.linspace(10, 500, n_data)

fig, ax = plt.subplots(figsize=(10, 10))

#ax.scatter(x, y, size of point)
ax.scatter(x_data, y_data, s=s_arr)
```

```python
import matplotlib.pyplot as plt
import numpy as np

n_data = 10

x_data = np.linspace(0, 10, n_data)
y_data = np.linspace(0, 10, n_data)

c_arr = [(c/n_data, c/n_data, c/n_data) for c in range(n_data)]

fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(x_data, y_data,s=500,c=c_arr)
```
