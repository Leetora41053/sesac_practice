# apply(*****)

```python
df_kospi = pd.read_csv('https://raw.githubusercontent.com/jin0choi1216/dataset/main/KOSPI_stocks.csv', index_col = 0)
df_kospi.head()
```

```python
# 방법 1
def check_change(data):
  if data > 0:
    return '오른주식'
  else:
    return '내린주식'
```

```python
# df_kospi['ChagesRatio'].apply(함수)
df_kospi['방법1'] = df_kospi['ChagesRatio'].apply(check_change)
df_kospi.head()
```

```python
# 방법 2
df_kospi['방법2'] = df_kospi['ChagesRatio'].apply(lambda x : '오른주식' if x > 0 else '내린주식')
df_kospi.head()
```

```python
import numpy as np
df_kospi['방법3'] = np.where(df_kospi['ChagesRatio'] > 0, '오른주식', '내린주식')
df_kospi.head()
```

### value_count()

```python
(df_kospi['방법1'] == df_kospi['방법2']).value_counts()
```

```python
(df_kospi['방법2'] == df_kospi['방법3']).value_counts()
```

## 실습
Q. 주당 순이익(EPS)이 0 이상 기업 중 PER 순위 - EPS 순위의 차이가 가장 적은 10개의 기업 출력
- PER는 낮을 수록 낮은 순위
- EPS는 높을 수록 낮은 순위


hint
- EPS가 0이상인 새로운 데이터프레임 선언
- EPS, PER의 순위(랭킹)를 새로운 칼럼 선언
- EPS의 순위칼럼 - PER의 순위칼럼을 연산한 새로운 칼럼
- EPS의 순위칼럼 - PER의 순위칼럼을 내림차순 정렬
- 상위 10개 기업 출력


```python
df = pd.read_csv('https://raw.githubusercontent.com/jin0choi1216/dataset/main/market_fundamental_20230817.csv', index_col = '티커')
df2 = df[df['EPS'] >= 0]
df2['EPS_rank'] = df2['EPS'].rank(ascending = False)
df2['PER_rank'] = df2['PER'].rank()
df2['PER_EPS_diff_rank'] =  abs(df2['PER_rank'] - df2['EPS_rank'])
df2.sort_values('PER_EPS_diff_rank').head(10)
```

# Merge(***)

```python
market_price_change = pd.read_csv('https://raw.githubusercontent.com/jin0choi1216/dataset/main/market_price_change_2022.csv')
purchases_of_equities = pd.read_csv('https://raw.githubusercontent.com/jin0choi1216/dataset/main/purchases_of_equities_2022.csv')
print(market_price_change.shape)
print(purchases_of_equities.shape)
```

```python
merge1 = pd.merge(left = market_price_change, right = purchases_of_equities, on = ['티커', '종목명'])
merge1.shape # how = 'inner'
```

```python
merge2 = pd.merge(left = market_price_change
                  , right = purchases_of_equities, on = ['티커', '종목명'], how = 'left')
merge2.shape
```

```python
merge3 = pd.merge(left = market_price_change
                  , right = purchases_of_equities, on = ['티커', '종목명'], how = 'right')
merge3.shape
```

```python
merge4 = pd.merge(left = market_price_change
                  , right = purchases_of_equities, on = ['티커', '종목명'], how = 'outer')
merge4.shape
```

```python
# 칼럼이름이 서로 다를 때
# 의도적으로
purchases_of_equities2 = purchases_of_equities.rename(
    columns = {"티커" : "이름변경_티커", '종목명' : '이름변경_종목명'})
```

```python
pd.merge(left = market_price_change, right = purchases_of_equities2,
         left_on = ['티커', '종목명'], right_on = ['이름변경_티커', '이름변경_종목명'], how = 'inner')
```

# pivot_table

```python
#group by랑 비슷한 역할을 한다
#웬만하면 group by를 쓰는 게 더 편하다

# 2012-01 ~ 2023-08

GOOGL = pd.read_csv('https://raw.githubusercontent.com/jin0choi1216/dataset/main/GOOGL.csv')
AAPL = pd.read_csv('https://raw.githubusercontent.com/jin0choi1216/dataset/main/AAPL.csv')
```

```python
# 한 데이터 프레임에 데이터가 두 번 잡힘
GOOGL['Symbol'] = 'GOOGLE'
AAPL['Symbol'] = 'AAPL'
df = pd.concat([GOOGL, AAPL])
df
```

```python
#인덱스랑 밸류를 재구조화 할 때 유리
pd.pivot_table(data = df, index = 'Date', columns = 'Symbol', values = 'Close')
```

```python
df[['Date', 'Symbol','Close']].groupby(['Date', 'Symbol']).mean()
```

```python
#aggfunc을 이용할 수 있다는 걸 인지
pd.pivot_table(data = df, index = ['Date', 'Symbol'], values = 'Close', aggfunc = 'mean')
```
