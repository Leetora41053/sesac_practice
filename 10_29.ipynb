```python
#P62

import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)

n_data = 100
random_noise1 = np.random.normal(0,1, ((n_data),))
random_noise2 = np.random.normal(1,1, ((n_data),))
random_noise3 = np.random.normal(2,1, ((n_data),))

fig, ax = plt.subplots(figsize=(10,7))
ax.tick_params(labelsize=30)

ax.plot(random_noise1,
        label='random noise1')
ax.plot(random_noise2,
        label='random noise2')
ax.plot(random_noise3,
        label='random noise3')

ax.legend()

ax.legend(fontsize=20)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

PI = np.pi
t = np.linspace(-4*PI, 4*PI, 300)
sin = np.sin(t)

fig, ax = plt.subplots(figsize=(10, 10))

for ax_idx in range(12):
  label_template = 'added by {}'
  ax.plot(t, sin+ax_idx, label=label_template.format(ax_idx))

ax.legend(fontsize=15, ncol=4, bbox_to_anchor=(0.5, -0.05), loc='upper center')

fig.tight_layout()
```

```python
import matplotlib.pyplot as plt
import numpy as np

PI = np.pi
t = np.linspace(-4*PI, 4*PI, 100)
sin = np.sin(t)

fig, ax = plt.subplots(figsize=(10, 10))

ax.plot(t, sin, label='sin(t)')

ax.plot(t, sin+1, marker='o', label='sin(t)+1', linestyle=':')

ax.plot(t, sin+2, marker='D', label='sin(t)+2', linestyle='--')

ax.plot(t, sin+3, marker='s', label='sin(t)+3', linestyle='-.')

ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=20)

fig.tight_layout()
```
